```python
from pathlib import Path
import warnings
import platform

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.compose import ColumnTransformer
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer
from sklearn.inspection import permutation_importance
from sklearn.metrics import (
    accuracy_score,
    balanced_accuracy_score,
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    RocCurveDisplay,
    silhouette_score,
    davies_bouldin_score,
    adjusted_rand_score,
    normalized_mutual_info_score,
)
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_validate
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.cluster import KMeans

RANDOM_STATE = 42
TEST_SIZE = 0.20

np.random.seed(RANDOM_STATE)
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', 100)
pd.set_option('display.width', 140)

print('Python:', platform.python_version())
```

```python
DATA_CANDIDATES = [
    Path('heart-disease-UCI.csv'),
#    Path('kidney_disease1.csv'),
#    Path('ChronicKidneyDisease.csv'),
    Path('/mnt/data/heart-disease-UCI.csv'),
#    Path('/mnt/data/kidney_disease1.csv'),
#    Path('/mnt/data/ChronicKidneyDisease.csv'),
]

def find_dataset(candidates):
    for path in candidates:
        if path.exists():
            return path
    raise FileNotFoundError(
        'Dataset not found. Upload kidney_disease.csv or update DATA_CANDIDATES with your file path.'
    )

data_path = find_dataset(DATA_CANDIDATES)
raw_df = pd.read_csv(data_path)
print('Loaded:', data_path)
print('Shape:', raw_df.shape)
display(raw_df.head())
display(raw_df.info())
```

```python
COLUMN_ALIASES = {
    'age': 'age',
    'sex': 'sex',
    'cp': 'chest_pain',
    'restbps': 'blood_pressure',
    'chol': 'serum_cholesterol',
    'fbs': 'fasting_blood_sugar',
    'restecg': 'resting_ECG',
    'thalach': 'heart_rate',
    'exang': 'exercise_angina',
    'oldpeak': 'ST_depression',
    'slope': 'slope',
    'ca': 'number_major_vessels',
    'thal': 'thal',
    'target': 'target',
}

clean_df = raw_df.copy()
#clean_df.columns = [str(c).strip().lower() for c in clean_df.columns]
clean_df = clean_df.rename(columns={c: COLUMN_ALIASES.get(c, c) for c in clean_df.columns})

TARGET = 'target'

NUMERIC_FEATURES = [
    'age', 
    'sex', 
    'chest_pain', 
    'blood_pressure',
    'serum_cholesterol',
    'fasting_blood_sugar',
    'resting_ECG',
    'heart_rate',
    'exercise_angina',
    'ST_depression',
    'slope',
    'number_major_vessels', 
    'thal',
    'target'
]
#CATEGORICAL_FEATURES = [
#    'red_blood_cells', 'pus_cell', 'pus_cell_clumps', 'bacteria',
#    'hypertension', 'diabetes_mellitus', 'coronary_artery_disease',
#    'appetite', 'peda_edema', 'anemia'
#]

FEATURES = NUMERIC_FEATURES

missing_cols = [c for c in [TARGET] + FEATURES if c not in clean_df.columns]
if missing_cols:
    raise ValueError(f'Missing required columns: {missing_cols}')

for col in NUMERIC_FEATURES:
    clean_df[col] = pd.to_numeric(clean_df[col], errors='coerce')

clean_df[TARGET] = (
    clean_df[TARGET]
    .astype(str)
    .str.strip()
    .str.lower()
#    .map({'ckd': 1, 'notckd': 0, 'not ckd': 0, '1': 1, '0': 0})
)

clean_df = clean_df.dropna(subset=[TARGET]).copy()
clean_df[TARGET] = clean_df[TARGET].astype(int)

print('Clean shape:', clean_df.shape)
print('Target distribution, 1 = CKD, 0 = Not CKD')
#display(clean_df[TARGET].value_counts().rename(index={1:'CKD', 0:'Not CKD'}))
display(clean_df.head())
```
